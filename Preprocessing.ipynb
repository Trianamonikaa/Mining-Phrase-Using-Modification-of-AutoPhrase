```python
from collections import Counter
from string import digits
# import inflect


def remove_number(term):
    for i in range(len(term)):
        if i is not i.isdigit():
            term = re.sub(r'^https?:\/\/.*[\r\n]*','', term,flags=re.MULTILINE)
    return term

def remove_stopwords(term):
    new_words = []
    for word in term:
        if word not in stopwords.words('indonesian'):
            new_words.append(word)
    return new_words
    
def remove_punctuation(term):
    new_words = []
    for word in term:
        new_word = re.sub(r'[^\w\s]', '', word)
        if new_word != '':
            new_words.append(new_word)
    return new_words
    
def case_folding(term):
    term = [words.lower() for words in term]
    return term


    

    
```
